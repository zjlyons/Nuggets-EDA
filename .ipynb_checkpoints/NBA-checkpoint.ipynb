```python
import pandas as pd
import numpy as np
import seaborn as sb
import json

from pandas.io.json import json_normalize

%matplotlib inline
import matplotlib.pyplot as plt
import altair as alt
alt.renderers.enable('notebook')

import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy.stats import boxcox
from fbprophet import Prophet
```

```python
from nba_api.stats.static import teams

nba_teams = teams.get_teams()
# Select the dictionary for the Celtics, which contains their team ID
nuggets = [team for team in nba_teams if team['abbreviation'] == 'DEN'][0]
nuggets_id = nuggets['id']
```

```python
from nba_api.stats.endpoints import leaguegamefinder

# Query for games where the Celtics were playing
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=nuggets_id)
# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]
games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])
games['year'] = games['GAME_DATE'].dt.year
games['month'] = games['GAME_DATE'].dt.month
games['months_of_data'] = (games['GAME_DATE'] - pd.Timestamp('1996-11-01'))/pd.Timedelta(1,'M')
games['rolling_plus_minus'] = games['PLUS_MINUS'].rolling(21).mean()
#games['points_boxcox'] = boxcox(games['PTS'])
games = games.dropna()
games
```

# Time Series Analysis

```python
data = list(games['PLUS_MINUS'].rolling(21).mean())
dates = list(games['GAME_DATE'])
plus_minus_s = pd.Series(data,index=dates)
plus_minus_s = plus_minus_s.fillna(0)
plus_minus_s.head()
```

```python
plus_minus_s.plot(figsize=(16,4))
```

```python
decomposition = sm.tsa.seasonal_decompose(games.set_index('GAME_DATE')['rolling_plus_minus'],model='additive',freq=82)

plt.rcParams["figure.figsize"] = (8,8)
f = decomposition.plot()
```

# De-Periodized Data

```python
pm_ts_df = games[['GAME_DATE','months_of_data','PLUS_MINUS']]
#pm_ts_df['rolling_plus_minus'] = pm_ts_df['PLUS_MINUS'].rolling(82).mean()
pm_ts_df = pm_ts_df.reset_index()
#pm_ts_df = pm_ts_df.set_index('GAME_DATE')
pm_ts_df
```

```python
dates = pd.DataFrame({'date':pd.date_range('1996-11-01','2019-03-18',freq='D')})
pm_time_series = dates.merge(pm_ts_df,how='left',left_on='date',right_on='GAME_DATE')
pm_time_series = pm_time_series.drop(columns=['index','GAME_DATE'])
pm_time_series['PLUS_MINUS']=pm_time_series['PLUS_MINUS'].fillna(0)
pm_time_series['months_of_data']=pm_time_series['months_of_data'].interpolate()
pm_time_series['month']=pm_time_series['date'].dt.month
pm_time_series = pm_time_series.set_index('date')
pm_time_series['rolling_plus_minus']=pm_time_series['PLUS_MINUS'].rolling(82).mean()
pm_time_series = pm_time_series.dropna()
pm_time_series.tail()
```

```python
g = sb.lmplot(x='months_of_data',y='rolling_plus_minus',data=pm_time_series,aspect=2.25,ci=0,
              scatter_kws={'color':'k','s':25},line_kws={'lw':5})

ax = g.axes[0,0]
```

```python
m1 = smf.ols(formula = 'rolling_plus_minus ~ months_of_data',data=pm_time_series).fit()
m1.summary()
```

```python
m2 = smf.ols(formula = 'rolling_plus_minus ~ months_of_data + C(month)',data=pm_time_series).fit()
m2.summary()
```

```python
predict_pm = pd.DataFrame({'date':pd.date_range('2017-01-01','2025-01-01',freq='D')},
                                  index=range(240,240+2923))

predict_pm['months_of_data'] = (predict_pm['date'] - pd.Timestamp('1996-11-01'))/pd.Timedelta(1,'M')
predict_pm['month'] = predict_pm['date'].apply(lambda x:x.month)

predict_pm.head()
```

```python
predict_pm['m1'] = m1.predict({'months_of_data':predict_pm['months_of_data']})
predict_pm['m2'] = m2.predict({'months_of_data':predict_pm['months_of_data'],'month':predict_pm['month']})

f,ax = plt.subplots(1,1,figsize=(16,8))

games.plot(x='GAME_DATE',y='rolling_plus_minus',ax=ax,c='k',label='Observations',lw=3,alpha=.5)
predict_pm.plot(x='date',y='m1',ax=ax,c='r',label='Simple',alpha=.5)
predict_pm.plot(x='date',y='m2',ax=ax,c='g',label='Fixed Effects',alpha=.5)
#predict_pm.plot(x='date',y='m2',ax=ax,c='b',label='Fixed effects',alpha=.5)
ax.set_xlim((pd.Timestamp('1996-11-01'),pd.Timestamp('2025-01-01')))
ax.set_ylim(-40,40)
ax.legend(loc='center left',bbox_to_anchor=(1,.5))
ax.set_xlabel('Date')
ax.set_ylabel('Plus-Minus')
```

# Prophet

```python
prophet_plus_minus = games[['GAME_DATE','rolling_plus_minus']].copy()
prophet_plus_minus.rename(columns={'GAME_DATE':'ds','rolling_plus_minus':'y'},inplace=True)
prophet_plus_minus.head()
```

```python
m3 = Prophet(weekly_seasonality=False,daily_seasonality=False).fit(prophet_plus_minus)
```

```python
m3_future = m3.make_future_dataframe(periods=60,freq='M')
m3_future.head()
```

```python
m3_forecast = m3.predict(m3_future)
m3_forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
f,ax = plt.subplots(1,1,figsize=(18,9))
_ = m3.plot(m3_forecast,ax=ax)

# Always label your axes
ax.set_xlabel('Time')
ax.set_ylabel('Plus/Minus')
ax.set_title('Observation and Forecast of Denver Nuggets Plus/Minus')
```

```python
f = m3.plot_components(m3_forecast)
```
